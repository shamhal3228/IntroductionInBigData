```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go

data = pd.read_csv('july-2018-temperature-precip.csv', sep=",")
print(data.head())
print('**********************space******************************')
#print(data.info())

fig = go.Figure(px.bar(data, x='date', y='max_temp', color = 'date', text='max_temp'))
fig.update_traces(marker=dict(line=dict(color='black', width=2)))
fig.update_layout(
    title={
        'text': "Температура Июля 2018 года",
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    title_font_size=20,
    xaxis_tickfont_size = 14,
    yaxis_tickfont_size = 14,
    width=None,
    height=700,
    margin=dict(l=0, r=0, t=35, b=0)
    )

fig.update_xaxes(
    tickangle = 315, 
    title_text = "Дата",
    title_font = {"size": 16}
    )
fig.update_yaxes(
    title_text = "Максимальная температура",
    title_font = {"size": 16}
    )
fig.show()

figPie = go.Figure(px.pie(data, values='max_temp', names='date', color = 'date'))
figPie.update_traces(marker=dict(line=dict(color='black', width=2)))
figPie.update_layout(
    title={
        'text': "Температура Июля 2018 года",
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
        xaxis_tickfont_size = 14,
        yaxis_tickfont_size = 14,
        width=None,
        height=700,
        margin=dict(l=0, r=0, t=35, b=0))
figPie.show()

figLine = go.Figure()
figLine.add_trace(go.Scatter(x=data['date'], y=data['max_temp'], mode='lines+markers', line_color='crimson', marker_color='white'))
figLine.update_traces(marker=dict(line=dict(color='black', width=2)))
figLine.update_xaxes(showgrid=True, gridwidth=2, gridcolor='ivory')
figLine.show()
```
